#### What is pytorch?
- [Pytorch](https://pytorch.org/) is an open source python-based deep learning library
- It is a machine learning framework used by researchers and developers for building and training deep learning models for NLP, vision, reinforcement learning applications
- PyTorch is dominant in research (over 70% of AI research implementations) and rapid prototyping ["PyTorch Grows as the Dominant Open Source Framework for AI and ML: 2024 Year in Review, Eli Uriegas", Meta and Jennifer Bly, Pytorch Blog](https://pytorch.org/blog/2024-year-in-review/#:~:text=Over%2070%25%20of%20AI%20research%20implementations%20are,25%25%2C%20enhancing%20both%20software%20and%20hardware%20capabilities.)
- However, as of 2025, TensorFlow still leads in overall adoption, commanding roughly 38% of the market share while PyTorch sits at 23% ["PyTorch vs TensorFlow: Which Deep Learning Framework Is Better?, Jason Karlin", AceCloud Blog](https://acecloud.ai/blog/pytorch-vs-tensorflow/#:~:text=While%20PyTorch%20is%20king%20in,while%20PyTorch%20sits%20at%2023%25.)
